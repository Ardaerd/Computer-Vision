```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("rainbow.jpg")
```

```python
plt.imshow(img)
```

```python
# Adding the 0 flag to read it in black and white
img = cv2.imread("rainbow.jpg",0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
# Binary Threshold
# if src(x,y) > thresh
#   dst(x,y) = maxValue
# else
#   dst(x,y) = 0
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
# Inverse Binary Threshold
# if src(x,y) > thresh
#   dst(x,y) = 0
# else
#   dst(x,y) = maxValue
    
ret, thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap='gray')
```

```python
# Truncate Threshold
# if src(x,y) > thresh
#   dst(x,y) = thresh
# else
#   dst(x,y) = src(x,y)


ret, thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)

plt.imshow(thresh3,cmap='gray')
```

```python
plt.imshow(img,cmap='gray')
```

```python
# Threshold to Zero
# if src(x,y) > thresh
#   dst(x,y) = src(x,y)
# else
#   dst(x,y) = 0

ret, thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh4,cmap='gray')
```

```python
# Inverted Threshold to Zero
# if src(x,y) > thresh
#   dst(x,y) = 0
# else
#   dst(x,y) = src(x,y)

ret, thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh5,cmap='gray')
```

```python
img = cv2.imread("crossword.jpg",0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(16,23))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,13,13)
show_pic(th2)
```

```python
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,14)
show_pic(th3)
```

```python
blended = cv2.addWeighted(src1=th1,alpha=0.7,src2=th2,beta=0.3,gamma=0)
show_pic(blended)
```
